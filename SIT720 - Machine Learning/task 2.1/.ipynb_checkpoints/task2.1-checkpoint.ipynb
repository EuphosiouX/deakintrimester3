```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
# Fetch the data 
df = pd.read_csv('microclimate-sensors-data.csv', delimiter=",")
```

```python
# Reference: https://stackoverflow.com/questions/52865323/how-to-count-missing-data-in-each-column-in-python
print('Feature name with numbers of missing entries:')
print(df.isnull().sum())
```

```python
replacement_values = {}

for column in df.columns:
    if df[column].isnull().sum() > 0:
        if pd.api.types.is_numeric_dtype(df[column]):
            # Check skewness to decide between mean and median
            skewness = df[column].skew()
            if abs(skewness) > 1:  # Highly skewed
                replacement = df[column].median()
                method = 'median'
            else:
                replacement = df[column].mean()
                method = 'mean'
            
            # Fixed: Avoid chained assignment warning
            df[column] = df[column].fillna(replacement)
            
        else:  # For categorical data
            replacement = df[column].mode()[0]
            df[column] = df[column].fillna(replacement)
            method = 'mode'
        
        replacement_values[column] = {'value': replacement, 'method': method}

# Print replacement values
print("\nReplacement Values for Missing Data:")
print("-" * 40)
for feature, info in replacement_values.items():
    if pd.api.types.is_numeric_dtype(df[feature]):
        print(f"{feature.ljust(25)}: {info['value']:.2f} (using {info['method']})")
    else:
        print(f"{feature.ljust(25)}: \"{info['value']}\" (using {info['method']})")
```

Because it accounts for all values and provides a balanced center estimate, the mean is most useful when the feature has a normal or symmetric distribution. However, if there are extreme values present, the result may be skewed because the mean is extremely sensitive to outliers. However, because it represents the middle value and is less vulnerable to the impact of extreme values, the median is a more robust estimator in situations when the data is skewed or contains outliers. However, neither the mean nor the median are suitable for categorical features. The mode, where the most frequent category is a meaningful representation of missing values, should to be used instead. This is because categorical data often cannot be meaningfully averaged or ordered.

```python
# Plot histograms for PM25 and PM10
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['PM25'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of PM2.5')
plt.xlabel('PM2.5 Concentration')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['PM10'], bins=10, kde=True, color='salmon')
plt.title('Distribution of PM10')
plt.xlabel('PM10 Concentration')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Show correlation
correlation = df[['PM25', 'PM10']].corr().iloc[0,1]
print(f"Correlation between PM2.5 and PM10: {correlation:.2f}")

# Print distribution statistics
print("\nDistribution Statistics:")
print("PM2.5:")
print(df['PM25'].describe())
print("\nPM10:")
print(df['PM10'].describe())
```

Both PM2.5 and PM10 have right-skewed distributions, according to the histograms, with a few extreme outliers and the majority of values grouped at the low end. With PM10 has a median of 6.0 g/m and a lower maximum of 308 g/m, PM2.5 has a median of 3.0 g/m and a maximum of over 1000 g/m. Even though they both measure air particles, there is extremely little correlation between them (0.05), meaning that changes in one do not always correspond to changes in the other. This means that they may come from various origins or show different weather behaviours.

```python
# Split 'LatLong' into 'Latitude' and 'Longitude'
df[['Latitude', 'Longitude']] = df['LatLong'].str.split(',', expand=True)
df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)

# Display encoded values
df[['LatLong', 'Latitude', 'Longitude']].head()
```

The LatLong string is divided into two separate numerical data, Latitude and Longitude, and converted to float values using theencoding technique. Since LatLong represents continuous geographic coordinates rather than categorical data, this approach is ideal. The true nature of the data would be distorted and geographical information would be lost if conventional categorical encoding techniques like one-hot or label encoding were used. The geographic coordinates can be effectively used for modeling, distance calculations, clustering, and spatial analyses by being separated and converted to numeric types. This method is the best encoding strategy for location data since it maintains the quantitative relationship between sites and allows the application of geospatial algorithms.

```python
continuous_features = [
    'MinimumWindDirection', 'AverageWindDirection', 'MaximumWindDirection',
    'MinimumWindSpeed', 'AverageWindSpeed', 'GustWindSpeed',
    'AirTemperature', 'RelativeHumidity', 'AtmosphericPressure',
    'PM25', 'PM10', 'Noise'
]

# Plot original distributions
plt.figure(figsize=(15, 10))
for i, feature in enumerate(continuous_features, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Original {feature}')
plt.tight_layout()
plt.suptitle('Original Distributions', y=1.02)
plt.show()
```

```python
# Apply Min-Max scaling
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[continuous_features])
df_scaled = pd.DataFrame(scaled_data, columns=continuous_features)

# Plot scaled distributions
plt.figure(figsize=(15, 10))
for i, feature in enumerate(continuous_features, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df_scaled[feature], kde=True)
    plt.title(f'Scaled {feature}')
plt.tight_layout()
plt.suptitle('Min-Max Scaled Distributions', y=1.02)
plt.show()

# Compare summary statistics
original_stats = df.describe().loc[['min', 'max', 'mean', 'std']]
scaled_stats = df_scaled.describe().loc[['min', 'max', 'mean', 'std']]

print("\nOriginal Statistics:")
print(original_stats)
print("\nScaled Statistics:")
print(scaled_stats)
```

Yes, there is a noticeable difference between the before and afterMin-Max scaling. Continuous characteristics like wind speeds, air temperature, humidity, and pollution concentrations (PM2.5, PM10) show general, varied scales in the original plots, with a few significant outliers extending the distribution and many values strongly skewed toward lower ranges. All feature values are standardized to a range between 0 and 1 following Min-Max scaling, but each distribution's shape, relative density,and skewness remains unchanged.  

For algorithms that are sensitive to feature magnitude (e.g., KNN, SVM, neural networks), this transformation is especially helpful because it guarantees that no variable will dominate others because of scale inequality (Ali, 2022). Standardizing the range allows for more balanced and efficient model training, even when the scaling doesn't change the underlying distribution.
